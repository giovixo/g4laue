# Test the Laue Lens beam model

```python
import pandas as pd
```

```python
df = pd.read_csv('output.txt', header=None, sep=' ')

fig, axes = plt.subplots()
x = df[0]
y = df[1]

axes.scatter(x, y, alpha=0.6, s=1)
axes.set_title('Uniform points in a ring (data from the g4 sim)')
axes.set_xlabel('x (cm)')
axes.set_ylabel('y (cm)')
axes.set_aspect('equal')
axes.grid(True)

plt.savefig('uniform_ring.png')
plt.show()
```
